```python
import numpy as np
import os
from PIL import Image

# npy
data = np.load('imgs_temp.npy')

# 
output_folder = r'C:\Users\xiaog\Desktop\Project\Datasets\PHOTOVOLTAIC THERMAL IMAGES DATASET\imgs_temp'

# 
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# data512x640
for idx, img_data in enumerate(data):
    #  512x640
    img_data = np.resize(img_data, (512, 640))

    #  0-255 
    img_data_normalized = np.uint8(255 * (img_data - np.min(img_data)) / (np.max(img_data) - np.min(img_data)))
    
    # NumPy
    image = Image.fromarray(img_data_normalized)
    
    # 
    image_path = os.path.join(output_folder, f'imgs_temp_{idx}.png')
    
    # 
    image.save(image_path)
    
    print(f"Image {idx} saved as {image_path}")

```

```python
import numpy as np
import os
from PIL import Image

#  mask 
mask_data = np.load('imgs_mask.npy')

# 
output_folder = r'C:\Users\xiaog\Desktop\Project\Datasets\PHOTOVOLTAIC THERMAL IMAGES DATASET\imgs_mask'

# 
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

#  mask  512x640 0  1
for idx, mask in enumerate(mask_data):
    #  mask  512x640
    mask = np.resize(mask, (512, 640))

    #  mask  0  1
    image = Image.fromarray(np.uint8(mask * 255))  # 0 -> , 1 -> 

    # 
    image_path = os.path.join(output_folder, f'imgs_mask_{idx}.png')

    #  mask 
    image.save(image_path)

    print(f"Mask {idx} saved as {image_path}")

```
