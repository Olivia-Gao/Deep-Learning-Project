```python
# inception
def inception_block(x, filter_1x1):
    # Firsr Branch
    conv1x1 = SeparableConv2D(filter_1x1, (1, 1), padding='same', activation='relu')(x)

    # Second Branch
    conv1x1_2 = SeparableConv2D(filter_1x1, (1,1), padding='same', activation='relu')(x)
    conv1x3 = SeparableConv2D(filter_1x1, (3,3), padding='same', activation='relu')(conv1x1_2)
   
    # Third Branch
    conv1x1_3 = SeparableConv2D(filter_1x1, (1,1), padding='same', activation='relu')(x)
    conv1x3_3 = SeparableConv2D(filter_1x1, (1,3), padding='same', activation='relu')(conv1x1_3)
    conv3x1 = SeparableConv2D(filter_1x1, (3,1), padding='same', activation='relu')(conv1x3_3)
    
    branch_4 = MaxPooling2D((3,3), strides=1, padding='same')(x)
    branch_4 = SeparableConv2D(filter_1x1, (1,1), padding='same', activation='relu')(branch_4)
    # Concatenate all outputs
    branch_output = concatenate([conv1x1, conv1x3, conv3x1,branch_4], axis=-1)

    return branch_output
```

```python
from tensorflow.keras import layers

# Attention Mechanism (Self-Attention Block)
class AttentionBlock(layers.Layer):
    def __init__(self, filters):
        super(AttentionBlock, self).__init__()
        self.filters = filters
        self.query_conv = SeparableConv2D(filters, (1, 1))
        self.key_conv = SeparableConv2D(filters, (1, 1))
        self.value_conv = SeparableConv2D(filters, (1, 1))
        self.attention_conv = SeparableConv2D(1, (1, 1), activation='sigmoid')

    def call(self, inputs):
        query = self.query_conv(inputs)
        key = self.key_conv(inputs)
        value = self.value_conv(inputs)

        # Compute attention weights
        attention_weights = tf.matmul(query, key, transpose_b=True)
        attention_weights = tf.nn.softmax(attention_weights, axis=-1)

        # Apply attention weights to the value
        attention_output = tf.matmul(attention_weights, value)
        attention_output = self.attention_conv(attention_output)

        return attention_output
```

```python
# Inception + Attention Model
from tensorflow.keras import models

# Inception + Attention + BiLSTM Model
def inception_attention_bilstm_model(input_shape, filter_1x1, attention_filters, lstm_units, num_classes):
    input_tensor = layers.Input(shape=input_shape)  # Example input shape (128, 128, 3)

    # Step 1: Apply Inception Block
    inception_out = inception_block(input_tensor, filter_1x1)
    
    # Step 2: Apply Attention Block to the Inception Block's output
    attention_out = AttentionBlock(attention_filters)(inception_out)

    # Step 3: Flatten and apply LSTM
    x = layers.GlobalAveragePooling2D()(attention_out)  # Reduce dimensions to (batch_size, features)
    x = layers.Reshape((1, -1))(x)  # Reshape for BiLSTM (batch_size, time_steps, features)

    # Add BiLSTM layer
    x = Bidirectional(LSTM(lstm_units))(x)

    # Step 4: Fully connected layers for classification
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dense(32, activation='relu')(x)
    output_tensor = layers.Dense(num_classes, activation='softmax')(x)

    # Build the model
    model = models.Model(inputs=input_tensor, outputs=output_tensor)
    
    return model
# 创建模型
model = inception_attention_model(input_shape = (640, 512, 3), filter_1x1=32, attention_filters=64, num_classes=3)

# 查看模型结构
model.summary()

```

```python

# 编译模型
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 训练模型
history = model.fit(
    train_generator,
    epochs=20,  # 训练10轮（可以根据需要调整）
    batch_size=32,
    validation_data=validation_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_steps=validation_generator.samples // validation_generator.batch_size
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Predictions for calculating additional metrics
y_pred = model.predict(test_generator, steps=len(test_generator))

# Since the model outputs softmax results (multi-class classification), we need to select the class with the highest probability for each sample
y_pred_classes = np.argmax(y_pred, axis=1)

# Get true labels
y_true = test_generator.classes  # If the generator returns integer class labels

# Classification report: precision, recall, f1-score, support
print(classification_report(y_true, y_pred_classes, zero_division=1))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1', 'Class 2'], yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# ROC Curve & AUC
# For multi-class tasks, we need to calculate the ROC curve and AUC for each class
n_classes = y_pred.shape[1]  # Number of classes
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true, y_pred[:, i], pos_label=i)
    roc_auc[i] = roc_auc_score(y_true == i, y_pred[:, i])  # Calculate AUC for each class

# Plot ROC curves for each class
plt.figure(figsize=(6, 6))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'ROC Class {i} (AUC = {roc_auc[i]:.2f})')
    
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()

# Precision-Recall Curve
# For multi-class classification, we need to compute the precision-recall curve for each class
plt.figure(figsize=(6, 6))
for i in range(n_classes):
    precision, recall, _ = precision_recall_curve(y_true == i, y_pred[:, i])
    plt.plot(recall, precision, label=f'Class {i} Precision-Recall Curve')

plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='lower left')
plt.show()

# F1-Score (weighted average for multi-class classification)
f1 = f1_score(y_true, y_pred_classes, average='weighted')  # 'weighted' calculates the weighted average F1 score for each class
print(f'Weighted F1-Score: {f1}')
```
