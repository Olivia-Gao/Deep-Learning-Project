```python
# Inception Block
def inception_block(x, filter_1x1):
    # First Branch
    conv1x1 = SeparableConv2D(filter_1x1, (1, 1), padding='same', activation='relu')(x)
    conv1x1 = BatchNormalization()(conv1x1)

    # Second Branch
    conv1x1_2 = SeparableConv2D(filter_1x1, (1, 1), padding='same', activation='relu')(x)
    conv1x3 = SeparableConv2D(filter_1x1, (3, 3), padding='same', activation='relu')(conv1x1_2)
    conv1x3 = BatchNormalization()(conv1x3)
   
    # Third Branch
    conv1x1_3 = SeparableConv2D(filter_1x1, (1, 1), padding='same', activation='relu')(x)
    conv1x3_3 = SeparableConv2D(filter_1x1, (1, 3), padding='same', activation='relu')(conv1x1_3)
    conv3x1 = SeparableConv2D(filter_1x1, (3, 1), padding='same', activation='relu')(conv1x3_3)
    conv3x1 = BatchNormalization()(conv3x1)
    
    # Fourth Branch (MaxPooling)
    branch_4 = MaxPooling2D((3, 3), strides=1, padding='same')(x)
    branch_4 = SeparableConv2D(filter_1x1, (1, 1), padding='same', activation='relu')(branch_4)
    branch_4 = BatchNormalization()(branch_4)
    
    # Concatenate all outputs
    branch_output = concatenate([conv1x1, conv1x3, conv3x1, branch_4], axis=-1)

    return branch_output
```

```python
def attention_module(x, reduction_ratio=8):
    channels = x.shape[-1]
    avg_pool = GlobalAveragePooling2D()(x)
    max_pool = GlobalMaxPooling2D()(x)
    
    avg_pool = Dense(channels // reduction_ratio, activation='relu')(avg_pool)
    avg_pool = Dense(channels, activation='sigmoid')(avg_pool)
    max_pool = Dense(channels // reduction_ratio, activation='relu')(max_pool)
    max_pool = Dense(channels, activation='sigmoid')(max_pool)
    
    channel_attention = Add()([avg_pool, max_pool])
    channel_attention = Activation('sigmoid')(channel_attention)
    channel_attention = Multiply()([x, channel_attention])

    avg_pool_spatial = tf.reduce_mean(channel_attention, axis=-1, keepdims=True)
    max_pool_spatial = tf.reduce_max(channel_attention, axis=-1, keepdims=True)
    spatial_attention = tf.concat([avg_pool_spatial, max_pool_spatial], axis=-1)
    spatial_attention = Conv2D(1, (7, 7), padding='same', activation='sigmoid')(spatial_attention)
    spatial_attention = Multiply()([channel_attention, spatial_attention])

    return spatial_attention
```

```python
from tensorflow.keras.layers import LSTM, Bidirectional

def bilstm_module(x, units=64):
  # 这里传递的是一个 LSTM 层实例
    x = Bidirectional(LSTM(units, return_sequences=True))(x)  # 修改为返回序列
    x = Bidirectional(LSTM(units, return_sequences=False))(x)  # 增加第二个 LSTM 层
    return x
```

```python
# 结合三个模块的完整模型
def model_with_inception_attention_bilstm(input_shape, filter_1x1, lstm_units=64, num_classes=2):
    input_layer = Input(shape=input_shape)

    # Inception + Attention 模块
    inception_attention_output = inception_block(input_layer, filter_1x1)
    attention_output = attention_module(inception_attention_output)

    # Global Average Pooling to reduce spatial dimensions
    gap_output = GlobalAveragePooling2D()(attention_output)

    # 将 Inception + Attention 输出 reshape 成 BiLSTM 输入格式
    reshaped_output = Reshape((-1, attention_output.shape[-1]))(gap_output)

    # BiLSTM 模块
    bilstm_output = bilstm_module(reshaped_output, lstm_units)

    # Dropout 防止过拟合
    x = Dropout(0.6)(bilstm_output)

    # 输出层
    output_layer = Dense(num_classes, activation='softmax')(x)  # 对于多分类任务，使用 softmax

    # 构建模型
    model = Model(inputs=input_layer, outputs=output_layer)
    
    return model

# 定义模型的输入形状
input_shape = (128, 128, 3)  # 假设输入图像的尺寸为 128x128，3 通道（RGB）
filter_1x1 = 64  # Inception 模块中的滤波器数量
lstm_units = 128 # BiLSTM 模块中的单元数
num_classes = 3  # 分类任务的类别数（例如：二分类）

# 创建模型
model = model_with_inception_attention_bilstm(input_shape, filter_1x1, lstm_units, num_classes)

# 查看模型的架构
model.summary()
```

```python
# 编译模型
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_generator, 
                    steps_per_epoch=len(train_generator), 
                    epochs=20, 
                    batch_size=32,
                    validation_data=test_generator,
                    validation_steps=len(test_generator)
                   )
```

```python
import matplotlib.pyplot as plt

# Visualize training and validation loss
plt.figure(figsize=(12, 6))

# Plot training and validation loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot training and validation accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# Evaluate on the test dataset
test_loss, test_accuracy = model.evaluate(test_generator, steps=len(test_generator))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Predictions for calculating additional metrics
y_pred = model.predict(test_generator, steps=len(test_generator))

# Since the model outputs softmax results (multi-class classification), we need to select the class with the highest probability for each sample
y_pred_classes = np.argmax(y_pred, axis=1)

# Get true labels
y_true = test_generator.classes  # If the generator returns integer class labels

# Classification report: precision, recall, f1-score, support
print(classification_report(y_true, y_pred_classes, zero_division=1))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1', 'Class 2'], yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# ROC Curve & AUC
# For multi-class tasks, we need to calculate the ROC curve and AUC for each class
n_classes = y_pred.shape[1]  # Number of classes
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true, y_pred[:, i], pos_label=i)
    roc_auc[i] = roc_auc_score(y_true == i, y_pred[:, i])  # Calculate AUC for each class

# Plot ROC curves for each class
plt.figure(figsize=(6, 6))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'ROC Class {i} (AUC = {roc_auc[i]:.2f})')
    
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()

# Precision-Recall Curve
# For multi-class classification, we need to compute the precision-recall curve for each class
plt.figure(figsize=(6, 6))
for i in range(n_classes):
    precision, recall, _ = precision_recall_curve(y_true == i, y_pred[:, i])
    plt.plot(recall, precision, label=f'Class {i} Precision-Recall Curve')

plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='lower left')
plt.show()

# F1-Score (weighted average for multi-class classification)
f1 = f1_score(y_true, y_pred_classes, average='weighted')  # 'weighted' calculates the weighted average F1 score for each class
print(f'Weighted F1-Score: {f1}')
```
