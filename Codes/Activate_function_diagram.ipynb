```python
import numpy as np
import matplotlib.pyplot as plt
import os

# 设置全局字体
plt.rc('font', family='Times New Roman', size=15)

# 激活函数定义
def sigmoid(x):
    return 1. / (1 + np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

def relu(x):
    return np.where(x < 0, 0, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))  # 减去最大值以防溢出
    return e_x / e_x.sum(axis=0)

# 绘图模板函数
def plot_function(x, y, title, save_path):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['left'].set_position(('data', 0))
    ax.spines['bottom'].set_position(('data', 0))
    ax.plot(x, y, color="black", lw=3)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.tight_layout()
    
    # 在图下方添加标题
    ax.text(0.5, -0.15, title, fontsize=16, family='Times New Roman',
            ha='center', va='top', transform=ax.transAxes)
    
    # 保存图像到指定路径
    if not os.path.exists(save_path):  # 创建保存路径（如果不存在的话）
        os.makedirs(save_path)
    
    file_path = os.path.join(save_path, f"{title.replace(' ', '_')}.png")
    plt.savefig(file_path)
    
    # 显示图像
    plt.show()
    plt.close(fig)

# 绘图函数分别调用
def plot_all():
    x = np.arange(-10, 10, 0.1)
    save_path = r'C:\Users\xiaog\Desktop\Project\Datasets\dataset\dataset 4\activate_function_diagram'
    
    # ReLU
    plot_function(x, relu(x), "ReLU Function", save_path)

    # Sigmoid
    plot_function(x, sigmoid(x), "Sigmoid Function", save_path)

    # Tanh
    plot_function(x, tanh(x), "Tanh Function", save_path)

# 调用主绘图函数
plot_all()
```
