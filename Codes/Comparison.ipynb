```python
import pickle
import matplotlib.pyplot as plt
import os
import numpy as np

# load history
def load_history(history_path):
    with open(history_path, 'rb') as f:
        history = pickle.load(f)
    return history

history_80_10_10 = load_history("models/history_split_80_10_10.pkl")
history_70_20_10 = load_history("models/history_split_70_20_10.pkl")
history_60_20_20 = load_history("models/history_split_60_20_20.pkl")

history_inception_70_20_10 = load_history("models/inception_model_history_split_70_20_10.pkl")
history_inception_attention_70_20_10 = load_history("models/inception_attention_model_history_split_70_20_10.pkl")
history_cnn_70_20_10 = load_history("models/cnn_model_split_70_20_10.pkl")
```

```python
# Define the directory to save the image
save_dir = r"C:\Users\xiaog\Desktop\Project\Datasets\dataset\Performance\Comparison"
os.makedirs(save_dir, exist_ok=True)  # Ensure the directory exists

plt.plot(history_80_10_10['accuracy'], label='Accuracy split 80-10-10')
plt.plot(history_70_20_10['accuracy'], label='Accuracy split 70-20-10')
plt.plot(history_60_20_20['accuracy'], label='Accuracy split 60-20-20')

plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Accuracy Comparison Across Datasets', fontsize=16)
plt.legend()
plt.grid(True)
plt.tight_layout()

accuracy_path = os.path.join(save_dir, 'accuracy_curve_comparison.png')
plt.savefig(accuracy_path, dpi=600, bbox_inches='tight')
print(f"Accuracy curve comparison saved at: {accuracy_path}")

plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Define the directory to save the image
save_dir = r"C:\Users\xiaog\Desktop\Project\Datasets\dataset\Performance\Comparison"
os.makedirs(save_dir, exist_ok=True)  # Ensure the directory exists

def compute_f1(precision, recall):
    precision = np.array(precision)
    recall = np.array(recall)
    return 2 * (precision * recall) / (precision + recall + 1e-8)

# Compute F1 scores
f1_70_20_10 = compute_f1(history_70_20_10['precision'], history_70_20_10['recall'])
f1_cnn = compute_f1(history_cnn_70_20_10['precision'], history_cnn_70_20_10['recall'])
f1_inception = compute_f1(history_inception_70_20_10['precision'], history_inception_70_20_10['recall'])
f1_inception_attention = compute_f1(history_inception_attention_70_20_10['precision'], history_inception_attention_70_20_10['recall'])

# Find minimum epoch length
min_len = min(len(f1_70_20_10), len(f1_cnn), len(f1_inception), len(f1_inception_attention))

# Trim all F1 scores to min_len
f1_70_20_10 = f1_70_20_10[:min_len]
f1_cnn = f1_cnn[:min_len]
f1_inception = f1_inception[:min_len]
f1_inception_attention = f1_inception_attention[:min_len]

# Epochs
epochs = range(min_len)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(epochs, f1_70_20_10, label='F1 Score - 70/20/10')
plt.plot(epochs, f1_cnn, label='CNN - 70/20/10')
plt.plot(epochs, f1_inception, label='Inception - 70/20/10')
plt.plot(epochs, f1_inception_attention, label='Inception + Attention - 70/20/10')

plt.title('F1 Score Comparison of Models (70/20/10 Split)', fontsize=16)
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('F1 Score', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save
F1_score_path = os.path.join(save_dir, 'F1_score_curve_comparison_models.png')
plt.savefig(F1_score_path, dpi=600, bbox_inches='tight')
print(f"F1_score curve comparison saved at: {F1_score_path}")

plt.show()
```
